```python
# Dependencies and Setup
import pandas as pd
import numpy as np

```

```python
first_cacao_df=pd.read_csv("flavors_of_cacao2.csv",encoding="utf-8")
first_cacao_df
```

```python
first_cacao_df.columns
```

```python
#drop bean column
no_bean_df=first_cacao_df.loc[:, ['Company (Maker-if known)', 'Specific Bean Origin or Bar Name', 'REF',
       'Review Date', 'Cocoa Percent', 'Company Location', 'Rating', 'Broad Bean Origin']]

no_bean_df

```

```python
#add na tp the blanks to drop them
no_bean_df["Broad Bean Origin"].replace("", np.nan, inplace=True)
no_bean_df
```

```python
#Drop any rows that don't contain all bean type & origin conditions. 
cacao_clean_df=no_bean_df.dropna()

cacao_clean_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
cacao_clean_df.to_csv("cacao_clean.csv", index=False, header=True)
```

```python
first_cacao_df["Bean Type"].replace(" ", np.nan, inplace=True)
first_cacao_df
```

```python
# Clean file with only dropna, keeping bean
cacaobean_dropna_df=first_cacao_df.dropna()
cacaobean_dropna_df
```
