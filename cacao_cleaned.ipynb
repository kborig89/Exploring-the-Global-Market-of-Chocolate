```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import csv_to_sqlite
```

```python
first_cacao_df=pd.read_csv("data/flavors_of_cacao2.csv",encoding="utf-8")
first_cacao_df
```

```python
first_cacao_df["Bean Type"].replace(" ", np.nan, inplace=True)
first_cacao_df["Broad Bean Origin"].replace(" ", np.nan, inplace=True)
first_cacao_df
```

```python

first_cacao_df = first_cacao_df.rename(columns={'Company (Maker-if known)': "company", 'Specific Bean Origin or Bar Name': "specific_bean_origin",
       'Review Date': "review_date", 'Cocoa Percent': "cocoa_percent", 'Company Location': "company_location", 'Rating': "rating",
       'Bean Type': "bean_type", 'Broad Bean Origin': "bean_origin_country"})
first_cacao_df
```

```python
first_cacao_df.columns
```

```python
#drop bean column
no_bean_df=first_cacao_df.loc[:, ['company', 'specific_bean_origin', 'review_date', 'cocoa_percent',
       'company_location', 'rating', 'bean_origin_country']]

no_bean_df

```

```python
#Drop any rows that don't contain all bean type & origin conditions. 
cacao_clean_df=no_bean_df.dropna()

cacao_clean_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
cacao_clean_df.to_csv("cacao_clean.csv", index=False, header=True)
```

```python
# Clean file with only dropna, keeping bean
cacaobean_dropna_df=first_cacao_df.dropna()
cacaobean_dropna_df
```

```python
#drop bean column
first_cacao_df=first_cacao_df.loc[:, ['company', 'specific_bean_origin', 'review_date', 'cocoa_percent',
       'company_location', 'rating', 'bean_type', 'bean_origin_country']]

first_cacao_df

```

```python
# Clean file with only dropna, keeping bean
cacaobean_dropna_df=first_cacao_df.dropna()
cacaobean_dropna_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
cacaobean_dropna_df.to_csv("cacao_clean_withbean.csv", index=False, header=True)
```

```python
import csv_to_sqlite 
# all the usual options are supported
options = csv_to_sqlite.CsvOptions(typing_style="full", encoding="utf-8") 
input_files = ["cacao_clean_withbean.csv"] # pass in a list of CSV files
csv_to_sqlite.write_csv(input_files, "cacao_bean.sqlite", options)
```

```python
conn = sqlite3.connect("cacao_bean.sqlite")
```

```python
cur = conn.cursor()
cur.execute("SELECT * FROM cacao_clean_withbean")
cur.fetchall()
```

```python

# create engine to connect to the ETL_Project DB
engine = create_engine('postgres://postgres:blU89bRd*@localhost:5432/cacao_db')
conn = engine.connect()

# add the cacao bean dataframe into the cacao table in postgres
cacaobean_dropna_df.to_sql('cacao_db', con = engine, if_exists = 'append', chunksize = 1000)

# return the cacao bean table
cacao = pd.read_sql("SELECT * FROM cacao_db", conn)
cacao.head()
```
