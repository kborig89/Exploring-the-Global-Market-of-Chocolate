```python
# Dependencies and Setup
import pandas as pd
import numpy as np

```

```python
first_cacao_df=pd.read_csv("data/flavors_of_cacao2.csv",encoding="utf-8")
first_cacao_df
```

```python
first_cacao_df["Bean Type"].replace(" ", np.nan, inplace=True)
first_cacao_df["Broad Bean Origin"].replace(" ", np.nan, inplace=True)
first_cacao_df
```

```python

first_cacao_df = first_cacao_df.rename(columns={'Company (Maker-if known)': "company", 'Specific Bean Origin or Bar Name': "specific_bean_origin",
       'Review Date': "review_date", 'Cocoa Percent': "cocoa_percent", 'Company Location': "company_location", 'Rating': "rating",
       'Bean Type': "bean_type", 'Broad Bean Origin': "bean_origin_country"})
first_cacao_df
```

```python
first_cacao_df.columns
```

```python
#drop bean column
no_bean_df=first_cacao_df.loc[:, ['company', 'specific_bean_origin', 'review_date', 'cocoa_percent',
       'company_location', 'rating', 'bean_origin_country']]

no_bean_df

```

```python
#Drop any rows that don't contain all bean type & origin conditions. 
cacao_clean_df=no_bean_df.dropna()

cacao_clean_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
cacao_clean_df.to_csv("cacao_clean.csv", index=False, header=True)
```
